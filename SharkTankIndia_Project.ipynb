In how many companies, each sharks invested

```python
col_names = ['Anupam','Ashneer','Namita','Aman','Peyush','Vineeta','Ghazal']
invest_by_shark = []
for col in col_names:
    a = 0
    for row in df[col]:
        if row == 'Y':
            a += 1
    invest_by_shark.append(a)
invest_by_shark
```

```python
plt.pie(invest_by_shark,labels = col_names,autopct='%1.1f%%')
plt.title('% of Total Compaines Invested by Shark')
plt.show()
```

Aman invested in 19.4%(highest in all sharks) companies
